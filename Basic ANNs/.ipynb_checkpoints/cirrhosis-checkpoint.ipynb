```python
df = df.dropna()
```

```python
def dummy(field, df):
    dummies = pd.get_dummies(df[field],drop_first=True)
    df = pd.concat([df.drop(field, axis=1), dummies], axis=1)    
    return df
```

```python
df = dummy(field='Drug', df=df)
```

```python
df
```

```python
df = dummy('Sex', df)
```

```python
df
```

```python
df = df.drop('Status', axis=1)
```

```python
df = dummy('Edema', df)
```

```python
df
```

```python
df['Ascites'].value_counts()
```

```python
df['Ascites']=df['Ascites'].apply(lambda Ascites: 1 if Ascites=='Y' else 0)
df['Hepatomegaly']=df['Hepatomegaly'].apply(lambda Hepatomegaly: 1 if Hepatomegaly=='Y' else 0)
df['Spiders']=df['Spiders'].apply(lambda Spiders: 1 if Spiders=='Y' else 0)
```

```python
df.columns = ['N_Days', 'Age', 'Ascites', 'Hepatomegaly', 'Spiders', 'Bilirubin',
       'Cholesterol', 'Albumin', 'Copper', 'Alk_Phos', 'SGOT', 'Tryglicerides',
       'Platelets', 'Prothrombin', 'Stage', 'Placebo', 'M', 'Edema_S', 'Edema_Y']
```

```python
X = df.drop('Stage', axis=1).values
y = df['Stage'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.losses import CategoricalCrossentropy
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(18, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(9, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(3, activation='softmax'))

model.compile(loss=CategoricalCrossentropy, optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, epochs=30,  validation_data=(X_test, y_test))
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

```python
predictions = model.predict_classes(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
```
