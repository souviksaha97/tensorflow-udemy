```python
def dummy(field, df):
    dummies = pd.get_dummies(df[field],drop_first=True)
    df = pd.concat([df.drop(field, axis=1), dummies], axis=1)    
    return df
```

```python
df = dummy(field='Drug', df=df)
```

```python
df = dummy('Sex', df)
```

```python
df = df.drop('Status', axis=1)
```

```python
df = dummy('Edema', df)
```

```python
df['Ascites']=df['Ascites'].apply(lambda Ascites: 1 if Ascites=='Y' else 0)
df['Hepatomegaly']=df['Hepatomegaly'].apply(lambda Hepatomegaly: 1 if Hepatomegaly=='Y' else 0)
df['Spiders']=df['Spiders'].apply(lambda Spiders: 1 if Spiders=='Y' else 0)
```

```python
X = df.drop('Stage', axis=1).values
y = df['Stage'].values
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
y_enc = LabelEncoder().fit_transform(y)
```

```python
import tensorflow
y_label = tensorflow.keras.utils.to_categorical(y_enc)
```

```python
y, y_label
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_label, test_size=0.5, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.losses import CategoricalCrossentropy
```

```python
X_train.shape
```

```python
y_test
```

```python
model = Sequential()
model.add(Dense( 18, input_dim=18, activation = 'relu'))
model.add(Dense( units = 9, activation= 'relu'))
model.add(Dense( units = 4, activation= 'softmax'))
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
```

```python
model.fit(x=X_train, y=y_train, epochs=500,  validation_data=(X_test, y_test), callbacks=[early_stop])
```
